```python
import matplotlib.pyplot as plt
import numpy
import netCDF4
```

```python
oht_980 = netCDF4.Dataset('data/ocean-heat-transport/oht_980.nc').variables['OHT_980'][0,:]
y25 = netCDF4.Dataset('data/ocean-heat-transport/oht_980.nc').variables['YQ'][:]
oht_986 = netCDF4.Dataset('data/ocean-heat-transport/oht_986.nc').variables['OHT_986'][0,:]
oht_987 = netCDF4.Dataset('data/ocean-heat-transport/oht_987.nc').variables['OHT_987'][0,:]
y5 = netCDF4.Dataset('data/ocean-heat-transport/oht_986.nc').variables['YQ1'][:]
oht_ncep = netCDF4.Dataset('data/ocean-heat-transport/ANNUAL_TRANSPORTS_1985_1989.nc').variables['OTn'][0,:]
y_ncep = netCDF4.Dataset('data/ocean-heat-transport/ANNUAL_TRANSPORTS_1985_1989.nc').variables['latitude'][:]
oht_gw = [0.6,1.8,-0.8,-0.6]
lat_gw = [47,24,-19,-30]
ohterr_gw = [0.1,0.3,0.6,0.3]
```

```python
plt.plot(y25, oht_980, 'k', label='OM4p25')
plt.plot(y5, oht_986, 'b', label='OM4p5')
plt.plot(y5, oht_987, 'r', label='OM4p5n')
plt.plot(y_ncep, oht_ncep, 'ko', label='NCEP')
plt.errorbar(lat_gw, oht_gw, yerr=ohterr_gw, fmt='m*', capsize=4, label='G&W')
plt.legend(loc='upper left');

plt.xlabel('Latitude ($^\circ$N)')
plt.ylabel('PW')
plt.title('Northward Global Ocean Heat Transport')
plt.xlim(-80,90)
plt.ylim(-2,2.2)
plt.grid()

plt.tight_layout()

plt.savefig('ocean-heat-transport-global.pdf')
```
